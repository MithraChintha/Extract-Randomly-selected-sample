```python
#filename="yellow_tripdata_2015-04.csv"
import glob 
filelist =glob.glob("*.csv") # get all the file names into list
```

```python
for file in filelist:
    n = sum(1 for line in open(file)) - 1 #number of records in file (excludes header)
    s=50000
    print(file)
    skip=sorted(random.sample(range(1,n+1),n-s))    
    df= pd.read_csv(file,sep=",",skiprows=skip) # read the 50000 rows of csv into df
    df.to_csv("sample_"+file, sep=',')  # save the file with suffix as sample_
    print("end")
```
